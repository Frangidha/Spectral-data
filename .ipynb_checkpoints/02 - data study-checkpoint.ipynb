## Handling of missing values

isna() & sum() are used to check if the amount of counted N/A in the colums and splitting the missing varaibles in catogorical and numerical list for imputation

```python
vars_missing_data = df_houseprice.columns[df_houseprice.isna().sum() > 0]
print("Variables with missing data:")
print(vars_missing_data) 

categorical_vars_with_missing_data = []
numerical_vars_with_missing_data = []

for var in vars_missing_data:
    if df_houseprice[var].dtype == 'object':
        categorical_vars_with_missing_data.append(var)
    else:
        numerical_vars_with_missing_data.append(var)
```

## Imputation of Categorical Data:

```python
print("Categorical Variables with Missing Data:")
categorical_vars_with_missing_data
```

the frequent imputation method was chosen It identifies the categorical variables with missing data, calculates the mode (most frequent category)

```python
from feature_engine.imputation import CategoricalImputer
categorical_imputer = CategoricalImputer(imputation_method='frequent', variables=categorical_vars_with_missing_data)
df_houseprice[categorical_vars_with_missing_data] = categorical_imputer.fit_transform(df_houseprice[categorical_vars_with_missing_data])
```

```python
print("Information of imputed categorical variables:")
df_houseprice[categorical_vars_with_missing_data].info()
```

# Correlation Study

change object into binaire numbers using onehotencoder

```python
from feature_engine.encoding import OneHotEncoder
encoder = OneHotEncoder(variables=df_houseprice.columns[df_houseprice.dtypes=='object'].to_list(), drop_last=False)
df_ohe = encoder.fit_transform(df_houseprice)
print(df_ohe.shape)
df_ohe.head(3)
```

We use .corr() for spearman and pearson methods, and investigate the top 10 correlations


- **Spearman Correlation**: Spearman correlation, on the other hand, is a non-parametric measure that assesses monotonic relationships between variables. It identifies whether variables tend to increase or decrease together, without assuming a constant rate of change. Spearman correlation is robust to outliers and works well with ordinal or ranked datred.


```python
corr_spearman = df_houseprice.corr(method='spearman')['SalePrice'].sort_values(key=abs, ascending=False)[1:].head(10)
corr_spearman
```

- **Pearson Correlation**: Pearson correlation is suitable for assessing linear relationships between variables. It quantifies the strength and direction of a linear relationship, ranging from -1 (perfect negative correlation) to 1 (perfect positive correlation). It is well-suited for data with linear associations.

```python
corr_pearson = df_houseprice.corr(method='pearson')['SalePrice'].sort_values(key=abs, ascending=False)[1:].head(10)
corr_pearson
```

```python
top_n = 5
important_variables = set(corr_pearson[:top_n].index.to_list() + corr_spearman[:top_n].index.to_list())
print("Variables to study:")
print(important_variables)
```

## Results
### Strong Correlations (0.7 and Higher)

- **OverallQual (0.809829)**: This attribute exhibits a very strong positive correlation with sale prices, indicating that higher overall quality is associated with higher sale prices.
- **GrLivArea (0.731310)**: The ground living area has a strong positive correlation with sale prices, implying that larger living spaces generally result in higher sale prices.

### Moderate Correlations (Between 0.5 and 0.7)

- **YearBuilt (0.652682)**: The year of construction shows a moderate positive correlation with sale prices, suggesting that newer houses tend to have higher sale prices.
- **GarageArea (0.649379)**: Garage area demonstrates a moderate positive correlation, implying that larger garages are associated with higher sale prices.
- **TotalBsmtSF (0.602725)**: The total basement area has a moderate positive correlation, indicating that larger basements are linked to higher sale prices.
- **GarageYrBlt (0.593788)**: The year the garage was built also displays a moderate positive correlation with sale prices.
- **1stFlrSF (0.575408)**: The area of the first floor has a moderate positive correlation, suggesting that larger first floors are connected to higher sale prices.

### Weak to Moderate Correlations (Between 0.5 and Lower)

- **YearRemodAdd (0.571159)**: The year of remodeling demonstrates a weak to moderate positive correlation, signifying that more recent remodeling tends to be associated with higher sale prices.
- **OpenPorchSF (0.477561)**: The open porch area exhibits a weak to moderate positive correlation, indicating that properties with larger open porches tend to have slightly higher sale prices.
- **LotArea (0.456461)**: Lot area has a weak to moderate positive correlation with sale prices, implying that larger lots may be associated with slightly higher sale prices.

# Visualization

```python
corr_matrix = df_houseprice.corr(method='pearson')
corr_with_saleprice = corr_matrix['SalePrice'].abs().sort_values(ascending=False)

# Select top correlated variables (excluding 'SalePrice' itself)
important_variables = corr_with_saleprice[1:7].index.tolist()

print("Important Variables:")
print(important_variables)
```

```python
corr_pearson = df_houseprice.corr(method='pearson')['SalePrice'].filter(important_variables)
corr_pearson_abs = corr_pearson.abs().sort_values(ascending=False)
top_correlations = corr_pearson_abs.head(6)
top_correlations_cols = top_correlations.index.tolist()

print("Top Correlated Variables:")
print(top_correlations_cols)
```

```python
correlation_data = df_houseprice[top_correlations_cols + ['SalePrice']].corr(method='pearson')

print("Correlation Coefficients:")
print(correlation_data)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Correlation Matrix

```python
def plot_heatmap(data):
    
    matrix = np.triu(correlation_data)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(data, annot=True, cmap='coolwarm', square=True, mask=matrix)
    plt.title("Correlation Matrix", fontsize=20)
    plt.show()

plot_heatmap(correlation_data)
```

```python
def plot_pairplot(data, variables):
    sns.pairplot(data[variables + [target_var]])
    plt.show()
    
target_var = 'SalePrice'
plot_pairplot(df_houseprice, important_variables)
```
